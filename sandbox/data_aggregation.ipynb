```python
def get_attr(df, city, idx_min, idx_max, attrib_name):
    attrib_data = df[df['GEO_NAME'] == city].reset_index().iloc[idx_min:idx_max].reset_index()
    new_df = pd.DataFrame()
    new_df[attrib_name] = attrib_data["DIM: Profile of Population Centres (2247)"]
    new_df["count"] = attrib_data["Dim: Sex (3): Member ID: [1]: Total - Sex"]
    return new_df

# van = get_attr(pc, "Vancouver", 1898, 1918, "labour_category")
# edm = get_attr(pc, "Edmonton", 1898, 1918, "labour_category")
tor = get_attr(pc, "Toronto", 1899, 1918, "labour_category")
tor
```
